# **DATA CLEANING**



```python
# drop unnecessary column's
df=df.drop(["App Id","Developer Id","Developer Website","Developer Email","Privacy Policy","Scraped Time"],axis=1)
df
```

```python
# drop column
df=df.drop("Minimum Android",axis=1)
```

```python
# check null values
df.isnull().sum()/len(df.index)*100
```

```python
df.info()
```

```python
# change datatype's
df["Released"]=pd.to_datetime(df["Released"], errors='coerce').dt.date
df["Free"]=df["Free"].astype("category")
df["Last Updated"]=pd.to_datetime(df["Last Updated"])
df["Ad Supported"]=df["Ad Supported"].astype("category")
df["In App Purchases"]=df["In App Purchases"].astype("category")
df["Editors Choice"]=df["Editors Choice"].astype("category")
```

```python
df.info()
```

```python
df=df.drop("Installs",axis=1)
df
```

```python
# drop null values
df=df[~df["App Name"].isnull()]
```

```python
df.isnull().sum()/len(df.index)*100
```

```python
df.shape
```

```python
# fill with 0
df["Rating"]=df["Rating"].fillna(df["Rating"][0])
df["Rating Count"]=df["Rating Count"].fillna(df["Rating Count"][0])
```

```python
df.isnull().sum()/len(df.index)*100
```

```python
# replace null values with "USD"
# replace "XXX" with "USD"
df["Currency"].fillna("USD",inplace=True)
df["Currency"].replace("XXX","USD",inplace=True)
```

```python
df.isnull().sum()/len(df.index)*100
```

```python
# replace with 0
df["Minimum Installs"]=df["Minimum Installs"].fillna(df["Minimum Installs"][0])
```

```python
df.isnull().sum()/len(df.index)*100
```

```python
# change datatype
df["Minimum Installs"]=df["Minimum Installs"].astype(int)
```

```python
df.info()
```

```python
# replace with mode
most_freq_date = df["Released"].mode()[0]
df["Released"].fillna(most_freq_date, inplace=True)
```

```python
df.isnull().sum()/len(df.index)*100
```

```python
# convert kb to mb and then replace with median
df["Size"] = df["Size"].str.replace("M", "", regex=True)
df["Size"] = df["Size"].str.replace("K", "", regex=True)
df["Size"] = pd.to_numeric(df["Size"], errors='coerce') / 1024
df["Size"].fillna(df["Size"].median(), inplace=True)
```

```python
df.head()
```

```python
# replace true and false with paid and free
df["Free"].replace({True: "Free", False: "Paid"}, inplace=True)
```

```python
df.isnull().sum()/len(df.index)*100
```

```python
# after cleaning the data
df.shape
```

## **EXPLORATORY DATA ANALYSIS**

```python
# 1) top 15 apps
top_20_apps=df["App Name"].value_counts().sort_values(ascending=False).head(15).reset_index()
plt.figure(figsize=(25,6))
sns.set_context("notebook",font_scale=1.2)
sns.set_style("whitegrid")
sns.barplot(data=top_20_apps,x="App Name",y="count",palette="Oranges_r",edgecolor="black")
plt.xlabel("App Name")
plt.ylabel("Count")
plt.title("Top 15 Most Popular Apps")
plt.tight_layout()
plt.show()
```

```python
# 2)top_15_app_category
top_category=df["Category"].value_counts().sort_values(ascending=False).reset_index().head(15)
plt.figure(figsize=(30,8))
sns.set_context("notebook",font_scale=1.4)
sns.set_style("whitegrid")
sns.barplot(data=top_category,x="Category",y="count",palette="viridis")
plt.xlabel("Category")
plt.ylabel("Count")
plt.title("Top 15 Category")
plt.tight_layout()
plt.show()
```

```python
# 3)free vs paid
free_paid=df["Free"].value_counts()
sns.reset_defaults()
colors=sns.color_palette("Set2")
plt.pie(free_paid.values,labels=free_paid.index,autopct="%1.2f%%",colors=colors)
plt.title("Free vs Paid")
plt.show()
```

```python
# 4)Average Installs by top 10 Category
avg_installs=df.groupby("Category")["Maximum Installs"].mean().reset_index().sort_values(by="Maximum Installs",ascending=False).head(10)
avg_installs
plt.figure(figsize=(12,5))
sns.reset_defaults()
sns.barplot(data=avg_installs,y="Category",x="Maximum Installs",palette="GnBu_r")
plt.ylabel("Category")
plt.xlabel("Average Installs")
plt.title("Average Installs by top 10 Category")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 5)average rating per category
avg_rating=df.groupby("Category")["Rating Count"].mean().reset_index().sort_values(by="Rating Count",ascending=False).head(10)
avg_rating
plt.figure(figsize=(10,5))
sns.barplot(data=avg_rating,y="Category",x="Rating Count",palette="Wistia_r",edgecolor="black")
plt.ylabel("Category")
plt.xlabel("Average Rating")
plt.title("Average Rating of top 10 Category")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 6)Top 10 categories with most Editors Choice Rating
editors_choice=df[df["Editors Choice"]==True]
editors_choice_count=editors_choice["Category"].value_counts().head(10).reset_index()
editors_choice_count.columns=["Category","count"]
plt.figure(figsize=(10,4))
sns.barplot(data=editors_choice_count,x="Category",y="count",palette="rocket")
plt.xlabel("Category")
plt.ylabel("Rating")
plt.title(" Top 10 categories with most Editors Choice Rating")
plt.tight_layout()
plt.show()
```

```python
# 7)distribute ad supported app by rating,paid and free
compare_rating=sns.FacetGrid(df,col="Ad Supported",col_wrap=3,height=4)
compare_rating.map(sns.barplot,"Rating","Free",palette="YlGn_r")
compare_rating.add_legend()
plt.show()
```

```python
# Extract year from date and then group by year and free
year=pd.to_datetime(df["Released"]).dt.year
print(year)
count_year=df.groupby([year,"Free"])["App Name"].count().reset_index()
count_year.columns=["Released","Free","count"]
print(count_year)
```

```python
# 10)ratio of Editors Choice
Editors_Choice_ratio=df["Editors Choice"].value_counts().sort_values(ascending=False).reset_index()
sns.barplot(data=Editors_Choice_ratio,x="Editors Choice",y="count")
plt.title("Ratio of Editors Choice")
plt.show()
```

```python
# 11)how editors choice apps scattered by rating count and maximum installs
sns.set_style("whitegrid")
sns.set_context("paper")
sns.scatterplot(data=df,x="Maximum Installs",y="Rating Count",hue="Editors Choice")
plt.show()
```

```python
# 12)Content rating
plt.figure(figsize=(8,4))
segment=df["Content Rating"].value_counts().sort_values(ascending=False).reset_index()
sns.set_context("paper")
sns.set_style("whitegrid")
sns.barplot(data=segment,x="Content Rating",y="count",palette="magma")
plt.title("Content Rating")
plt.show()
```

```python
# 13)editors choice by content ratings
content_rating_choice=sns.FacetGrid(df,col="Content Rating",col_wrap=6,height=4)
content_rating_choice.map(sns.barplot,"Editors Choice","Rating",palette="YlOrRd_r")
content_rating_choice.add_legend()
plt.show()
```

```python
# 14)Aggregate installs by Content Rating and Free
content_rating_choice_df = (
    df.groupby(["Content Rating", "Free"])["Maximum Installs"]
    .mean()
    .reset_index())
content_rating_choice = sns.FacetGrid(
    content_rating_choice_df, col="Content Rating", col_wrap=6, height=4, sharex=False)
content_rating_choice.map_dataframe(
    sns.barplot, x="Free", y="Maximum Installs", palette="viridis")
content_rating_choice.set_titles("{col_name}")
content_rating_choice.set_axis_labels("Free App", "Average Installs")
content_rating_choice.fig.suptitle(
    "Average Installs for Free vs Paid Apps by Content Rating", fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# 15)crosstabulation in Editors choice vs Free
cross_tab=pd.crosstab(df["Editors Choice"],df["Free"])
print(cross_tab)
plt.figure(figsize=(6,4))
sns.heatmap(cross_tab,annot=True,fmt="d",cmap="YlGnBu",linewidths=0.8)
plt.title("crosstabulation in Editors choice vs Free")
plt.xlabel("Free")
plt.ylabel("Editors Choice")
plt.show()
```

```python
# 16)Correlation matrix
numeric_df=df.select_dtypes(include="number")
corr=numeric_df.corr()
plt.figure(figsize=(6,4))
sns.heatmap(corr,annot=True,fmt=".2f",cmap="copper_r",linewidths=0.8)
plt.title("Correlation Matrix")
plt.show()
```

```python
# extract only paid apps details details
app_price=df[df["Price"]>0]
app_price
```

```python
# create visualization
plt.figure(figsize=(18,6))
sns.boxplot(data=app_price,x="Category",y="Price",hue="Editors Choice",palette="rocket")
plt.xlabel("Category")
plt.ylabel("Price")
plt.title("Boxplot of Price by Category and Editors Choice")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# calculate outliers
Q1=app_price["Price"].quantile(0.25)
Q3=app_price["Price"].quantile(0.75)
print("Q1:",Q1)
print("Q3:",Q3)
IQR=Q3-Q1
print("IQR:",IQR)
lower_bound=Q1-1.5*IQR
upper_bound=Q3+1.5*IQR
print("lower_bound:",lower_bound)
print("upper_bound:",upper_bound)
```

```python
# visualize outliers
outliers_find=app_price[(app_price["Price"]<lower_bound)|(app_price["Price"]>upper_bound)]
plt.figure(figsize=(18,6))
sns.boxplot(data=outliers_find,x="Category",y="Price",hue="Editors Choice",palette="rocket")
plt.xlabel("Category")
plt.ylabel("Price")
plt.title("Boxplot of Price by Category and Editors Choice")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# remove outliers
outliers=app_price[(app_price["Price"]>=lower_bound)&(app_price["Price"]<=upper_bound)]
```

```python
# 17)after remove ouliers
plt.figure(figsize=(20,6))
sns.boxplot(data=outliers,x="Category",y="Price",hue="Editors Choice",palette="rocket")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
